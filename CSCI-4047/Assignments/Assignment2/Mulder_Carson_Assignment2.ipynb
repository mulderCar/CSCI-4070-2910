```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('./Income Data.csv')
```

3. For *Income*, *Education*, and *Prestige* calculate and display the following data:
    - Minimum Value
    - 1st Quartile
    - Median
    - Mean
    - 3rd Quartile
    - Maximum Value

```python
income_min = np.min(df['income'])
income_q1 = np.percentile(df['income'], 25)
income_median = np.percentile(df['income'], 50)
income_mean = np.mean(df['income'])
income_q3 = np.percentile(df['income'], 75)
income_max = np.max(df['income'])

print(f'Income Min: {income_min}')
print(f'Income Q1: {income_q1}')
print(f'Income Median: {income_median}')
print(f'Income Mean: {income_mean}')
print(f'Income Q3: {income_q3}')
print(f'Income Max: {income_max}')

edu_min = np.min(df['education'])
edu_q1 = np.percentile(df['education'], 25)
edu_median = np.percentile(df['education'], 50)
edu_mean = np.mean(df['education'])
edu_q3 = np.percentile(df['education'], 75)
edu_max = np.max(df['education'])

print(f'Education Min: {edu_min}')
print(f'Education Q1: {edu_q1}')
print(f'Education Median: {edu_median}')
print(f'Education Mean: {edu_mean}')
print(f'Education Q3: {edu_q3}')
print(f'Education Max: {edu_max}')

prestige_min = np.min(df['prestige'])
prestige_q1 = np.percentile(df['prestige'], 25)
prestige_median = np.percentile(df['prestige'], 50)
prestige_mean = np.mean(df['prestige'])
prestige_q3 = np.percentile(df['prestige'], 75)
prestige_max = np.max(df['prestige'])

print(f'Prestige Min: {prestige_min}')
print(f'Prestige Q1: {prestige_q1}')
print(f'Prestige Median: {prestige_median}')
print(f'Prestige Mean: {prestige_mean}')
print(f'Prestige Q3: {prestige_q3}')
print(f'Prestige Max: {prestige_max}')
```

4. Plot a *histogram* for **prestige**
5. Plot a *histogram* for **education**

```python
plt.hist(df.prestige, bins = 12, color = 'orange', alpha = 0.5)
plt.title('Prestige Histogram')
plt.xlabel('Prestige')
plt.ylabel('Frequency')
plt.show()

plt.hist(df.education, bins = 12, color = 'green', alpha = 0.5)
plt.title('Education Histogram')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.show()
```

6. Plot a *scatter plot* for **income** and **education**
7. Plot a *scatter plot* for **education** and **prestige**

```python
plt.scatter(df.income, df.education, color = 'red', alpha = 0.5)
plt.title('Income vs Education')
plt.xlabel('Income')
plt.ylabel('Education')
plt.show()

plt.scatter(df.education, df.prestige, color = 'blue', alpha = 0.5)
plt.title('Education vs Prestige')
plt.xlabel('Education')
plt.ylabel('Prestige')
plt.show()
```

8. Calculate and display the *variance* and *standard deviation* for **income**


```python
income_variance = np.var(df['income'])
income_standard_deviation = np.std(df['income'])

print(f'Income Variance: {income_variance}')
print(f'Income Standard Deviation: {income_standard_deviation}')
```
