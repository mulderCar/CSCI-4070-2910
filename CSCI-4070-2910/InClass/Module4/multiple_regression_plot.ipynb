```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# sample data
np.random.seed(0)
x1 = np.random.rand(50)
x2 = np.random.rand(50)
y = 2 * x1 + 3 * x2 + np.random.rand(50)

coefficients = [2,3]

# create a 3d plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")

# plot the data points
ax.scatter(x1,x2,y,c='b', marker='o', label='Data Points')

# create a grid of points that plots the regression plane
# (not a line any more due to 3 dimensional space)
x1_range = np.linspace(min(x1), max(x1), 50)
x2_range = np.linspace(min(x2), max(x2), 50)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
# calculates predicted points
y_pred = coefficients[0] * x1_grid + coefficients[1] * x2_grid 

# plot the regression plane
ax.plot_surface(x1_grid, x1_grid, y_pred, alpha=0.5, cmap='viridis', label = "Regression Plane")

# labels and legend
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.show()
```
